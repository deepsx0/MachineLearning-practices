```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report , confusion_matrix
```

```python
from sklearn.linear_model import Perceptron
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
data = pd.DataFrame(iris.data , columns = iris.feature_names)
y = pd.DataFrame(iris.target , columns = ["variety"])
```

```python
std_scaler = StandardScaler()
```

```python
x_train , x_test ,y_train ,y_test = train_test_split(data , y)
```

```python
std_xtrain = std_scaler.fit_transform(x_train)
std_xtest = std_scaler.fit_transform(x_test)
```

```python
perceptron = Perceptron(max_iter = 50 , eta0 = 0.15 , tol = 1e-3)
perceptron.fit(std_xtrain , np.ravel(y_train))
```

```python
y_pred = perceptron.predict(std_xtest)
```

```python
print(np.ravel(y_test))
```

```python
print(np.ravel(y_pred))
```

```python
print(classification_report(y_pred , y_test))
```
