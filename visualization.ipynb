```python
from sklearn.tree import DecisionTreeClassifier
```

```python
x = samdf.iloc[:,:4]
```

```python
y = samdf.iloc[:,4]
```

```python
y.head()
```

```python
x_train , x_test , y_train , y_test = train_test_split(x , y, test_size = 0.3) 
```

```python
from sklearn import metrics
```

```python
accuracy = metrics.accuracy_score(y_test,y_predict)
print(accuracy)
```
